### Introduction to Classification

We are importing scikit-learn to load the built-in Iris dataset.
The Iris dataset contains measurements of 150 iris flowers from 3 different species: 
   - Setosa
   - Versicolor
   - Virginica

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

### Loading the iris dataset

iris.data: A 2D array (X) with 150 rows and 4 columns. Each row is a flower, and each column is a feature (e.g., sepal length, sepal width, petal length, petal width).

iris.target: The labels (y) representing the species of the flowers. It contains integers: 0, 1, or 2 corresponding to the three species.

```python
iris = load_iris()
X = iris.data
y = iris.target
```

### Splitting the Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

The above function helps us split our dataset into training data (used to build the model) and test data (used to evaluate it)
- test_size=0.3: 30% of the data will be used for testing, and 70% for training.

- random_state=42: Ensures reproducibility. Same data gets split the same way every time we run it.

### Creating the Decision Tree


We're creating an instance of a Decision Tree classifier, a machine learning model that splits the data into branches based on feature values to classify it.

```python

clf = DecisionTreeClassifier()
```

### Training the model

This is where the model learns. It analyzes the training data and creates a tree-like structure to best classify the data based on the features and labels

```python
clf.fit(X_train, y_train)
```

### Making predictions on the test dataset

This is where the model learns. It analyzes the training data and creates a tree-like structure to best classify the data based on the features and labels

```python
y_pred = clf.predict(X_test)
```

We pass the unseen test data (X_test) to the trained model, and it returns predictions (y_pred) about which class each test sample belongs to.

### Calculating accuracy

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

Compares the model’s predictions (y_pred) with the true labels (y_test) and calculates the accuracy—the percentage of correct predictions.

accuracy * 100:.2f% just formats it to 2 decimal places for neat display.

### Visualizations

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Set the size of the plot
plt.figure(figsize=(12, 8))

# Plot the decision tree
plot_tree(
    clf, 
    filled=True,                  # Color nodes based on class
    feature_names=iris.feature_names,  # Names of the input features
    class_names=iris.target_names      # Names of the target classes
)

# Show the plot
plt.title("Decision Tree for Iris Dataset")
plt.show()
```

Explanation of the tree diagram

- Feature-based splits (e.g., petal length (cm) <= 2.45)

- Gini impurity (a measure of purity in a node)

- Class distributions at each node

- Predicted class at the bottom of each node
